```python
import pandas as pd
import numpy as np
import csv
import os
from datetime import datetime
```

```python
path = '/Volumes/Cisco/Fall2021/onnx-exchange/conversion/'
path_output = '/Volumes/Cisco/Summer2022/onnx-exchange/analysis/conversion/'
```

```python
models = ['lenet5', 'resnet18', 'vgg', 'lstm', 'gru']
frameworks = ['Keras', 'pytorch']
```

```python
data_file = open(path_output+'traiing_validation_metrics_keras.csv', mode='w', newline='',
                                  encoding='utf-8')
data_writer = csv.writer(data_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
#data_writer.writerow(['model','Epoch', 'Training_accuracy', 'Validation_Accuracy', "Training_Loss", 'Validation_Loss'])
data_writer.writerow(['framework','model', 'inference_before', 'inference_after_onnx', 'inference_after_coreml', 'memory_before', 'memory_after_onnx', 'memory_after_coreml'])

for frame_ in frameworks:
    for model in models: 
        data_inference_before = []
        data_inference_after = {}

        data_memory_before = []
        data_memory_after = {}
        if os.path.exists(path+'errors/{}/{}'.format(frame_,model)):
            list_files = [x for x in os.listdir(path+'errors/{}/{}'.format(frame_,model)) if 'runtime_metrics_' in x]
            for i in range(1, 11):
                for file_ in list_files:
                    df = pd.read_csv(path+'errors/{}/{}/{}'.format(frame_,model, file_))
                    batch_size = df.batch_size.values.tolist()
                    converter = df.converter.values.tolist()
                    original_infererence_time = df.original_infererence_time.values.tolist()
                    converted_infererence_time = df.converted_infererence_time.values.tolist()
                    original_size = df.original_size.values.tolist()
                    converted_size = df.converted_size.values.tolist()
                    for j in range(len(converter)):
                        if batch_size[j] == 128:
                            data_inference_before.append(original_infererence_time[j])
                            data_memory_before.append(original_size[j])
                            if converter[j] in data_inference_after.keys():
                                #data_inference_before[converter[j]].append(original_infererence_time[j])
                                data_inference_after[converter[j]].append(converted_infererence_time[j])

                                #data_memory_before[converter[j]].append(original_size[j])
                                data_memory_after[converter[j]].append(converted_size[j])
                            else:
                                #data_inference_before[converter[j]] = [original_infererence_time[j]]
                                data_inference_after[converter[j]] = [converted_infererence_time[j]]

                                #data_memory_before[converter[j]] = [original_size[j]]
                                data_memory_after[converter[j]] = [converted_size[j]]
            #for key, val in data_inference_before.items()
            data_writer.writerow([frame_,model, np.mean(data_inference_before), np.mean(data_inference_after['onnx']), np.mean(data_inference_after['coremltools']), np.mean(data_memory_before), np.mean(data_memory_after['onnx']), np.mean(data_memory_after['coremltools'])])

data_file.close()
```

```python
import math

def convert_size(size_bytes): 
    if size_bytes == 0: 
        return "0B" 
    size_name = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    s = round(size_bytes / p, 2)
    #return "%s %s" % (s, size_name[i])
    return s
```

```python
data_file = open(path_output+'error_metrics_summary.csv', mode='w', newline='', encoding='utf-8')
data_writer = csv.writer(data_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
#data_writer.writerow(['model','Epoch', 'Training_accuracy', 'Validation_Accuracy', "Training_Loss", 'Validation_Loss'])
data_writer.writerow(['framework','model', 'abs_error_onnx', 'rel_error_onnx', 'abs_error_coreml', 'rel_error_coreml','', 'abs_std_onnx', 'rel_std_onnx', 'abs_std_coreml', 'rel_std_coreml'])

for frame_ in frameworks:
    for model in models: 
        data_abs_error = {}
        data_rel_error = {}
        if os.path.exists(path+'errors/{}/{}'.format(frame_,model)):
            list_files = [x for x in os.listdir(path+'errors/{}/{}'.format(frame_,model)) if 'runtime_metrics_' in x]
            for i in range(1, 11):
                for file_ in list_files:
                    df = pd.read_csv(path+'errors/{}/{}/{}'.format(frame_,model, file_))
                    batch_size = df.batch_size.values.tolist()
                    converter = df.converter.values.tolist()
                    avg_abs_error = df.avg_abs_error.values.tolist()
                    avg_rel_error = df.avg_rel_error.values.tolist()
                    for j in range(len(converter)):
                        if batch_size[j] == 128:
                            if converter[j] in data_abs_error.keys():
                                #data_inference_before[converter[j]].append(original_infererence_time[j])
                                data_abs_error[converter[j]].append(avg_abs_error[j])

                                #data_memory_before[converter[j]].append(original_size[j])
                                data_rel_error[converter[j]].append(avg_rel_error[j])
                            else:
                                #data_inference_before[converter[j]] = [original_infererence_time[j]]
                                data_abs_error[converter[j]] = [avg_rel_error[j]]

                                #data_memory_before[converter[j]] = [original_size[j]]
                                data_rel_error[converter[j]] = [avg_rel_error[j]]
            #for key, val in data_inference_before.items()
            data_writer.writerow([frame_, model, np.mean(data_abs_error['onnx']), np.mean(data_rel_error['onnx']), np.mean(data_abs_error['coremltools']), np.mean(data_rel_error['coremltools']),'', np.std(data_abs_error['onnx']), np.std(data_rel_error['onnx']), np.std(data_abs_error['coremltools']), np.std(data_rel_error['coremltools'])])
data_file.close()
```
